-Nh·∫≠n x√©t chung

C·∫£ hai ƒë·ªì th·ªã ƒë·ªÅu cho th·∫•y m·ªëi quan h·ªá ti√™u c·ª±c gi·ªØa gi√° b√°n v√† nƒÉm ƒëƒÉng k√Ω, s·ªë km ƒë√£ ƒëi. ƒêi·ªÅu n√†y c√≥ nghƒ©a l√† gi√° b√°n xe th∆∞·ªùng gi·∫£m d·∫ßn theo nƒÉm ƒëƒÉng k√Ω v√† s·ªë km ƒë√£ ƒëi.

```python
from scipy.stats import pearsonr
numeric_variables = dataset_numerical[variables_related_to_price]
```

```python
# T√≠nh to√°n h·ªá s·ªë t∆∞∆°ng quan v√† p-value
correlation_results = pd.DataFrame(columns=['Bi·∫øn s·ªë', 'H·ªá s·ªë t∆∞∆°ng quan Pearson', 'P-value'])
for col in numeric_variables.columns:
    correlation, p_value = pearsonr(numeric_variables[col], df['gi√°_b√°n'])
    correlation_results = correlation_results.append({'Bi·∫øn s·ªë': col, 'H·ªá s·ªë t∆∞∆°ng quan Pearson': correlation, 'P-value': p_value}, ignore_index=True)
```

```python
# Hi·ªÉn th·ªã k·∫øt qu·∫£
for index, row in correlation_results.iterrows():
    variable_name = row['Bi·∫øn s·ªë']
    correlation = row['H·ªá s·ªë t∆∞∆°ng quan Pearson']
    p_value = row['P-value']

    print(f"\nBi·∫øn s·ªë: {variable_name}")
    print(f"H·ªá s·ªë t∆∞∆°ng quan Pearson: {correlation}")
    print(f"P-value: {p_value}")

    if abs(correlation) >= 0.8:
        print("ƒê·ªìng thu·∫≠n m·∫°nh")
    elif abs(correlation) >= 0.5:
        print("ƒê·ªìng thu·∫≠n v·ª´a")
    elif abs(correlation) >= 0.3:
        print("ƒê·ªìng thu·∫≠n y·∫øu")
    else:
        print("Kh√¥ng ƒë·ªìng thu·∫≠n")

    if p_value < 0.001:
        print("P-value < 0.001: M·∫°nh")
    elif p_value < 0.05:
        print("P-value < 0.05: V·ª´a ph·∫£i")
    elif p_value < 0.1:
        print("P-value < 0.1: Y·∫øu")
    else:
        print("P-value > 0.1: Kh√¥ng c√≥ ƒë·ªß b·∫±ng ch·ª©ng th·ªëng k√™ ƒë·ªÉ k·∫øt lu·∫≠n.")

    # Nh·∫≠n x√©t v·ªÅ ·∫£nh h∆∞·ªüng
    if abs(correlation) >= 0.3 and p_value < 0.1:
        print("C√≥ ·∫£nh h∆∞·ªüng ƒë√°ng k·ªÉ ƒë·ªëi v·ªõi gi√° xe.")
    else:
        print("Kh√¥ng c√≥ ·∫£nh h∆∞·ªüng ƒë√°ng k·ªÉ ƒë·ªëi v·ªõi gi√° xe.")
```

# Ph√¢n t√≠ch ƒëa bi·∫øn

  - Ph√¢n t√≠ch ANOVA
  
  - Bi·ªÉu ƒë·ªì box plot

```python
# Ph√¢n t√≠ch ANOVA
from scipy.stats import f_oneway

categorical_columns = ['h√£ng_xe', 'd√≤ng_xe', 'lo·∫°i_xe', 'dung_t√≠ch_xe', 'xu·∫•t_x·ª©']
target_variable = 'gi√°_b√°n'

# T·∫°o m·ªôt t·ª´ ƒëi·ªÉn tr·ªëng ƒë·ªÉ l∆∞u k·∫øt qu·∫£ ANOVA v√† nh·∫≠n x√©t
ket_qua_anova = {}

# L·∫∑p qua t·ª´ng bi·∫øn ph√¢n lo·∫°i
for column in categorical_columns:
    # Th·ª±c hi·ªán ANOVA m·ªôt chi·ªÅu
    cac_nhom = [df[target_variable][df[column] == category] for category in df[column].unique()]
    f_statistic, p_value = f_oneway(*cac_nhom)

    # L∆∞u k·∫øt qu·∫£ ANOVA v√† nh·∫≠n x√©t v√†o t·ª´ ƒëi·ªÉn
    ket_qua_anova[column] = {'F-statistic': f_statistic, 'p-value': p_value}

    # Th√™m nh·∫≠n x√©t
    if p_value < 0.05:
        ket_qua_anova[column]['Nh·∫≠n x√©t'] = f"C√≥ ·∫£nh h∆∞·ªüng ƒë√°ng k·ªÉ ƒë·ªëi v·ªõi {target_variable}"
    else:
        ket_qua_anova[column]['Nh·∫≠n x√©t'] = f"Kh√¥ng c√≥ ·∫£nh h∆∞·ªüng ƒë√°ng k·ªÉ ƒë·ªëi v·ªõi {target_variable}"

# Hi·ªÉn th·ªã k·∫øt qu·∫£
for column, result in ket_qua_anova.items():
    print(f"\nK·∫øt qu·∫£ ANOVA cho {column}:\n{result}")
```

- H√£ng Xe (h√£ng_xe): F-test = 15.04, p-value = 2.8084e-17

F-test r·∫•t cao v√† p-value r·∫•t th·∫•p, cho th·∫•y c√≥ m·ªëi quan h·ªá m·∫°nh m·∫Ω gi·ªØa h√£ng xe v√† gi√° xe. C√≥ nghƒ©a l√† h√£ng xe l√† m·ªôt y·∫øu t·ªë quan tr·ªçng ·∫£nh h∆∞·ªüng ƒë·∫øn gi√° xe m√°y.

- D√≤ng Xe (d√≤ng_xe): F-test = 4.33, p-value = 1.786e-09

# **X√¢y d·ª•ng m√¥ h√¨nh**

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import sklearn.metrics

# T√°ch d·ªØ li·ªáu
X = df.drop(columns=["gi√°_b√°n"])
y = df["gi√°_b√°n"]

# X·ª≠ l√Ω c√°c gi√° tr·ªã string
le = OneHotEncoder()
X_encoded = le.fit_transform(X)

# X·ª≠ l√Ω bi·∫øn y
y_encoded = le.fit_transform(y.apply(lambda x: pd.Series([x])))

# Ph√¢n t√°ch d·ªØ li·ªáu
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded.toarray(), test_size=0.25)

# X√¢y d·ª±ng m√¥ h√¨nh
model = LinearRegression()

# Hu·∫•n luy·ªán m√¥ h√¨nh
model.fit(X_train, y_train)

# ƒê√°nh gi√° m√¥ h√¨nh
y_pred = model.predict(X_test)

# T√≠nh to√°n sai s·ªë b√¨nh ph∆∞∆°ng trung b√¨nh
mse = sklearn.metrics.mean_squared_error(y_test, y_pred)

# T√≠nh to√°n h·ªá s·ªë x√°c ƒë·ªãnh
r2 = sklearn.metrics.r2_score(y_test, y_pred)

# V·∫Ω bi·ªÉu ƒë·ªì ph√¢n ph·ªëi gi√° th·ª±c t·∫ø v√† d·ª± ƒëo√°n
plt.scatter(y_test, y_pred)
plt.xlabel("Gi√° th·ª±c t·∫ø")
plt.ylabel("Gi√° d·ª± ƒëo√°n")
plt.show()

# V·∫Ω bi·ªÉu ƒë·ªì ph√¢n b·ªë sai s·ªë
plt.hist(y_test - y_pred)
plt.xlabel("Sai s·ªë")
plt.show()

print("MSE: ", mse)
print("R2: ", r2)
```
